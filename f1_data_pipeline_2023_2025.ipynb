## Importing fastf1 and enabling cache for faster sessino load

```python
import fastf1
fastf1.Cache.enable_cache('/content/f1_cache')

import pandas as pd
import numpy as np
```

## Getting race data for Round 1 of the 2023 season

```python
session = fastf1.get_session(2023, 'Bahrain', 'Race')
session.load()

#One session contains several dataframes (results, laps, weather_data, car_data, track_status, etc.)
```

```python
session.results
```

```python
laps = session.laps
laps.head()
```

## Filtering unwanted attributes

```python
laps_selected = laps[[
    'Driver',
    'Team',
    'LapNumber',
    'LapTime',
    'Position',
    'Compound',
    'Stint'
]].copy()
```

```python
laps_selected['LapTimeSeconds'] = (
    laps_selected['LapTime']
    .dt.total_seconds()
)
```

## Adding metadata

```python
laps_selected['Season'] = 2023
laps_selected['Round'] = 1
laps_selected['RaceName'] = session.event['EventName']
```

```python
laps_selected.head()
```

## Output as .csv

```python
laps_selected.to_csv(
    '/content/output/race_laps_2023_round_1.csv',
    index=False
)
```

```python
from fastf1 import get_event_schedule

schedule = get_event_schedule(2023)
```

## Repeating above process for all rounds in 2023

```python
all_laps = []

# Load full race calendar and explicitly exclude testing
schedule = fastf1.get_event_schedule(2023)
schedule = schedule[schedule['RoundNumber'] > 0]

for _, event in schedule.iterrows():
    round_no = int(event['RoundNumber'])
    race_name = event['EventName']

    print(f"Processing Round {round_no}: {race_name}")

    try:
        session = fastf1.get_session(2023, round_no, 'Race')
        session.load()

        laps = session.laps.loc[:, [
            'Driver', 'Team', 'LapNumber', 'LapTime',
            'Position', 'Compound', 'Stint'
        ]].copy()

        # Convert lap time
        laps['LapTimeSeconds'] = laps['LapTime'].dt.total_seconds()

        # Metadata
        laps['Season'] = 2023
        laps['Round'] = round_no
        laps['RaceName'] = race_name

        all_laps.append(laps)

        print(f"✓ Completed {race_name}")

    except Exception as e:
        print(f"✗ Failed {race_name} — {e}")
        continue

# ---------- FINAL CONSOLIDATION ----------
final_df = pd.concat(all_laps, ignore_index=True)

print("\nSummary check:")
print("Races processed:", final_df['RaceName'].nunique())
print(final_df['RaceName'].value_counts())

# Export
final_df.to_csv("laps_2023.csv", index=False)
```

```python
quali_laps = []

for _, event in schedule.iterrows():
    try:
        session = fastf1.get_session(2023, event['RoundNumber'], 'Qualifying')
        session.load()

        result = session.results[[
            'DriverNumber', 'Abbreviation', 'FullName',
            'TeamName', 'Position', 'Q1', 'Q2', 'Q3'
        ]].copy()

        for col in ['Q1', 'Q2', 'Q3']:
            if col in result.columns:
                result[col] = result[col].dt.total_seconds()

        result['Position'] = pd.to_numeric(result['Position'], errors='coerce')
        result['Season'] = 2023
        result['Round'] = event['RoundNumber']
        result['RaceName'] = event['EventName']

        quali_laps.append(result)

        print(f"Processed: {event['EventName']}")

    except Exception as e:
        print(f"Skipped {event['EventName']}: {e}")
```

```python
quali_df_2023 = pd.concat(quali_laps, ignore_index=True)

quali_df_2023.to_csv(
    '/content/output/quali_laps_2023.csv',
    index=False
)
```
